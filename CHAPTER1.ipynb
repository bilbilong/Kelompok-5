Karena sifatnya, machine learning harus berusaha semaksimal mungkin untuk mendapatkan model yang sangat baik dari data pelatihan. Namun, model yang bekerja dari data pelatihan mungkin tidak mencerminkan data lapangan yang benar. Ini tidak berarti bahwa harus membuat model yang kurang akurat dari data pelatihan dengan sengaja. Hal ini akan merusak strategi fundamental dari machine learning. 

### **Menghadapi Overfitting**

Overfitting mempengaruhi tingkat kinerja *Machine Learning*. Ada dua metode umum yang digunakan untuk mengatasi overfitting yaitu :
1. Regulasi, merupakan metode numerik yang mencoba membangun struktur model sesederhana dengan biaya kinerja yang kecil. Model (kurva) yang kompleks cenderung overfitting. Kurva sederhana mencerminkan karakteristik keseluruhan kelompok dengan lebih baik.
2. Validasi adalah proses yang menyimpan sebagian dari data pelatihan dan menggunakannya untuk membantu kinerja. Kita dapat mengatakan bahwa model tersebut overfitting ketika model yang dilatih menghasilkan tingkat kinerja yang rendah untuk input data yang disediakan.



**Gambar 1.9** Membagi data pelatihan untuk proses validasi

Langkah-langkah proses validasi :
1. Bagi data peatihan menjadi dua kelompok yaitu untuk pelatihan dan untuk validasi. Rasio pelatihan dengan set validasi adalag 8:2
2. Latih model dengan set pelatihan
3. Evaluasi kinerja model menggunakan set validasi
a). Jika model menghasilkan kinerja yang memuaskan, selesaikan pelatihan
b). Jika performa tidak memberikan hasil memadai, modifikasi model dan ulangi proses dari  langkah 2.

Validasi silang merupakan sedikit variasi dari proses validasi yang membagi data pelatihan ke dalam beberapa kelompok pelatihan dan validasi, tetapi terus mengubah set data. Validasi silang mengulangi pembagian data , hal ini dilakukan  agar model dapat terlalu pas bahkan untuk set validasi ketika sudah diperbaiki. Validasi silang dapat mendeteksi overfitting model dengan lebih baik.

##### **Ringkasan**
- Kecerdasan Buatan, Mechine Learning, dan Deep LLarning berbeda. Namun ketiganya terkait satu sama lain dengan cara " Deep Learning adalah sejenis Machine Learning, dan Machine Learning adalah  sejenis Kecerdasan Buatan.
- Machine Learning adalah pendekatan induktif yang mendapatkan model dari data pelatihan yang berguna untuk pengenalan gambar, suara, dan pemrosesan bahasa alami, dll.
- Keberhasilan Machine Learning bergantung seberapa baik proses generalisasi diimplementasi. Untuk mencegah penurunan kinerja karena perbedaan antara data pelatihan dan data input aktual.
- Overfitting terjadi ketika model terlalu disesuaikan dengan data pelatihan sehingga menghasilkan kinerja yang buruk untuk data input yang sebenarnya. Overfitting adalah salah satu faktor utama yang mengurangii kinerja generalisasi.

- Bergantung pada metode pelatihan, Machine Learning dapat berupa,supervised learning, unsupervised learning, dan pembelajaran penguatan.
Supervised learning dibagi menjadi klasifikasi dan regresi, tergantung pada model.  Klasifikasi menentukan ke dalam kelompok mana data input termasuk. Regresi memprediksi nilai dan mengambil nilai untuk output yang benar dalam data latihan.
