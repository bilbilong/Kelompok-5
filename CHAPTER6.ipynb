# **Lapisan Konvolusi**

Bagian ini menjelaskan bagaimana lapisan konvolusi bekerja, yang merupakan salah satu sisi dari jaringan saraf ekstraksi fitur. Lapisan pooling, sisi lainnya, akan diperkenalkan pada bagian berikutnya.

Lapisan konvolusi menghasilkan gambar baru yang disebut feature maps. Feature map menonjolkan fitur unik dari gambar asli. Lapisan konvolusi bekerja dengan cara yang sangat berbeda dibandingkan lapisan jaringan saraf lainnya. Lapisan ini tidak menggunakan bobot koneksi dan penjumlahan berbobot. Sebagai gantinya, lapisan ini berisi filter yang mengubah gambar. Filter ini disebut filter konvolusi. Proses memasukkan gambar melalui filter konvolusi menghasilkan feature map.

Gambar 6-4 menunjukkan proses lapisan konvolusi, di mana tanda bintang (*) menandakan operasi konvolusi, dan tanda Ï† menandakan fungsi aktivasi. Ikon persegi abu-abu di antara operator ini menunjukkan filter konvolusi. Lapisan konvolusi menghasilkan jumlah feature maps yang sama dengan jumlah filter konvolusi. Oleh karena itu, misalnya, jika lapisan konvolusi berisi empat filter, maka akan dihasilkan empat feature maps.

Operasi konvolusi merupakan penjumlahan dari hasil perkalian elemen-elemen yang berada pada posisi yang sama di kedua matriks. Hasil sebesar 7 pada Gambar 6-6 dihitung sebagai berikut:

$$

Gambar 6-8. Operasi konvolusi ketiga

Setelah baris atas selesai, baris berikutnya dimulai lagi dari sebelah kiri (lihat Gambar 6-9).



Gambar 6-9. Operasi konvolusi dimulai lagi dari kiri

Proses tersebut diulang sampai peta fitur dari filter yang diberikan dihasilkan, seperti yang ditunjukkan pada Gambar 6-10.

````python
import numpy as np

# matriks citra (input image)
image = np.array([
    [1, 1, 1, 3],
    [4, 6, 4, 8],
    [30, 0, 1, 5],
    [0, 2, 2, 4]
])

# filter konvolusi
kernel = np.array([
    [1, 0],
    [0, 1]
])

# ukuran output (valid convolution)
output_shape = (image.shape[0] - kernel.shape[0] + 1,
                image.shape[1] - kernel.shape[1] + 1)

# inisialisasi feature map
feature_map = np.zeros(output_shape, dtype=int)

# operasi konvolusi
for i in range(output_shape[0]):
    for j in range(output_shape[1]):
        sub_matrix = image[i:i+kernel.shape[0], j:j+kernel.shape[1]]
        feature_map[i, j] = np.sum(sub_matrix * kernel)

print("Input Image:\n", image)
print("Kernel:\n", kernel)
print("Feature Map (Output):\n", feature_map)
````

# Lapisan Penggabungan
Lapisan pooling mengurangi ukuran gambar dengan menggabungkan piksel-piksel tetangga di area tertentu dari gambar menjadi satu nilai representatif. Pooling adalah teknik yang umum digunakan dan telah diterapkan dalam banyak skema pemrosesan gambar lainnya.

Untuk melakukan operasi di lapisan pooling, kita perlu menentukan cara memilih piksel pooling dari gambar dan cara menetapkan nilai representatif. Piksel tetangga biasanya dipilih dari matriks persegi, dan jumlah piksel yang digabungkan bervariasi tergantung pada masalah yang dihadapi. Nilai representatif biasanya ditetapkan sebagai rata-rata atau maksimum dari piksel yang dipilih. 

Operasi lapisan pooling ternyata cukup sederhana. Karena ini adalah operasi dua dimensi, dan penjelasan tertulis mungkin menimbulkan kebingungan, mari kita lihat contohnya. Pertimbangkan gambar input berukuran 4x4 piksel, yang diwakili oleh matriks yang ditunjukkan pada Gambar **6-15.**

**Gambar 6-16.** *Hasil penggabungan menggunakan dua metode yang berbeda*

Sebenarnya, dalam konteks matematis, proses pooling merupakan jenis operasi konvolusi. Perbedaannya dengan lapisan konvolusi adalah bahwa filter konvolusi bersifat statis, dan area konvolusi tidak tumpang tindih. Contoh yang diberikan pada bagian berikutnya akan menjelaskan hal ini lebih lanjut. Lapisan pooling dapat mengkompensasi objek yang tidak berpusat dan miring hingga batas tertentu. Misalnya, lapisan pooling dapat meningkatkan pengenalan kucing yang mungkin tidak berada di tengah gambar input. Selain itu, karena proses pooling mengurangi ukuran gambar, hal ini sangat bermanfaat untuk mengurangi beban komputasi dan mencegah overfitting.

Fungsi MnistConv, yang melatih jaringan menggunakan algoritma back-propagation,

[W1, W5, WO] = MnistConv(W1, W5, WO, D)

Di mana W1, W5, dan Wo masing-masing merupakan matriks filter konvolusi, matriks bobot lapisan tersembunyi pooling, dan matriks bobot lapisan keluaran tersembunyi. X dan D masing-masing merupakan input dan keluaran yang benar dari data pelatihan. Daftar berikut menunjukkan berkas MnistConv.m, yang mengimplementasikan fungsi MnistConv.

Daftar berikut menunjukkan fungsi Conv, yang dipanggil oleh fungsi MnistConv
. Fungsi ini mengambil citra masukan dan matriks filter konvolusi, lalu
mengembalikan peta fitur. Kode ini terdapat dalam berkas Conv.m.

````python

def Conv(x, W):
    """
    x : numpy array (HxWxC_in) atau (HxW) kalau single channel
    W : numpy array (wrow x wcol x numFilters)
    return: y (yrow x ycol x numFilters)
    """
    wrow, wcol, numFilters = W.shape
    xrow, xcol = x.shape[:2]

    yrow = xrow - wrow + 1
    ycol = xcol - wcol + 1
    y = np.zeros((yrow, ycol, numFilters))

    for k in range(numFilters):
        filter_k = W[:, :, k]
        # rotasi 180 derajat seperti rot90(...,2) di MATLAB
        filter_rot = np.rot90(filter_k, 2)
        y[:, :, k] = convolve2d(x[:, :], filter_rot, mode='valid')
    
    return y
````

Fungsi MnistConv juga memanggil fungsi Pool, yang diimplementasikan
dalam daftar berikut. Fungsi ini mengambil peta fitur dan mengembalikan gambar
setelah proses penggabungan rata-rata 2 x 2Â´. Fungsi ini terdapat dalam berkas Pool.m.

````python


def Pool(x):
    """
    2x2 mean pooling (stride 2)

    x : numpy array (HxWxC)  -> feature maps
    return: y (H/2 x W/2 x C)
    """
    xrow, xcol, numFilters = x.shape
    y = np.zeros((xrow // 2, xcol // 2, numFilters))

    filter_mean = np.ones((2, 2)) / 4.0   # 2x2 mean filter

    for k in range(numFilters):
        image = convolve2d(x[:, :, k], filter_mean, mode='valid')
        # ambil tiap 2 langkah (downsampling stride 2)
        y[:, :, k] = image[::2, ::2]

    return y
````

Ada sesuatu yang menarik tentang kode ini; ia memanggil fungsi konvolusi dua dimensi, conv2, persis seperti fungsi Conv. Ini karena proses pooling adalah salah satu jenis operasi konvolusi. Rata-rata pooling dari contoh ini diimplementasikan menggunakan operasi konvolusi dengan filter berikut:

3. Ekstraktor fitur ConvNet terdiri dari tumpukan bergantian antara lapisan konvolusi dan lapisan pooling. Karena ConvNet berhubungan dengan citra dua dimensi, sebagian besar operasinya dilakukan dalam bidang konseptual dua dimensi.

4. Dengan menggunakan filter konvolusi, lapisan konvolusi menghasilkan citra yang menonjolkan fitur dari citra masukan. Jumlah citra keluaran dari lapisan ini sama dengan jumlah filter konvolusi yang dimiliki jaringan. Filter konvolusi sebenarnya hanyalah matriks dua dimensi.

5. Lapisan pooling mengurangi ukuran citra. Ia mengikat piksel tetangga dan menggantinya dengan nilai representatif. Nilai representatif ini bisa berupa nilai maksimum atau nilai rata-rata dari piksel.
