## **Node pada Jaringan Saraf**

Setiap kali kita mempelajari sesuatu, otak kita menyimpan pengetahuan tersebut. Komputer juga menggunakan memori untuk menyimpan informasi. Meskipun keduanya sama-sama menyimpan informasi, mekanismenya sangat berbeda. Komputer menyimpan informasi pada lokasi tertentu dalam memori, sedangkan otak mengubah asosiasi antar-neuron. Neuron itu sendiri tidak memiliki kemampuan penyimpanan; ia hanya mengirimkan sinyal dari satu neuron ke neuron lainnya. Otak merupakan sebuah jaringan raksasa yang terdiri dari neuron-neuron ini, dan asosiasi antar-neuron membentuk informasi tertentu.

Jaringan saraf meniru mekanisme otak. Sama seperti otak yang tersusun dari hubungan banyak neuron, jaringan saraf dibangun dengan hubungan antar-node, yaitu elemen yang merepresentasikan neuron pada otak. Jaringan saraf meniru asosiasi antar-neuron, yang merupakan mekanisme terpenting dari otak, dengan menggunakan nilai bobot (weight value). Tabel berikut merangkum analogi antara otak dan jaringan saraf.

![Tabel 2-1.png](<attachment:Tabel 2-1.png>)

Menjelaskan hal ini lebih jauh dengan teks saja mungkin akan menimbulkan kebingungan. Perhatikan sebuah contoh sederhana untuk pemahaman yang lebih baik tentang mekanisme jaringan saraf. Misalkan sebuah node menerima tiga masukan (input), seperti yang ditunjukkan pada Gambar 2-2.

![Gambar 2-2.png](<attachment:Gambar 2-2.png>)

**Gambar 2-2.** Sebuah node yang menerima tiga masukan

Lingkaran dan panah pada gambar menunjukkan node dan aliran sinyal. $x_1,x_2,x_3$ adalah sinyal masukan. $w_1,w_2,w_3$ adalah bobot untuk sinyal masing-masing. Terakhir, $b$ adalah bias, yaitu faktor lain yang berhubungan dengan penyimpanan informasi. Dengan kata lain, informasi dalam jaringan saraf disimpan dalam bentuk bobot (weights) dan bias.

Sinyal masukan dari luar dikalikan dengan bobot sebelum mencapai node. Setelah sinyal berbobot tersebut terkumpul pada node, nilai-nilai ini dijumlahkan menjadi jumlah berbobot (weighted sum). Jumlah berbobot pada contoh ini dihitung sebagai berikut:

$$v = (w_1 \times x_1) + (w_2 \times x_2) + (w_3 \times x_3) + b$$

Persamaan ini menunjukkan bahwa sinyal dengan bobot yang lebih besar akan memiliki pengaruh yang lebih besar. Sebagai contoh, jika bobot $w_1 = 1$ dan $w_2 = 5$ maka sinyal $x_2$ memiliki pengaruh lima kali lebih besar dibandingkan dengan $x_1$ ketika $w_1 = 0$, maka $x_1$ sama sekali tidak ditransmisikan ke node. Ini berarti $x_1$ terputus dari node. Contoh ini menunjukkan bahwa bobot dalam jaringan saraf meniru cara otak mengubah asosiasi antar-neuron.

Persamaan jumlah berbobot dapat dituliskan dengan matriks sebagai berikut:

$$v = wx + b$$

di mana $w$ dan $x$ didefinisikan sebagai berikut:

$$
w = \begin{bmatrix} w_1 & w_2 & w_3 \end{bmatrix}, \quad
x = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
$$

Akhirnya, node memasukkan jumlah berbobot ke dalam fungsi aktivasi dan menghasilkan keluarannya. Fungsi aktivasi menentukan perilaku dari node tersebut.

$$
y = \varphi(v)
$$

Dari persamaan ini, $Ï†(.)$ adalah fungsi aktivasi. Ada banyak jenis fungsi aktivasi yang tersedia dalam jaringan saraf, dan kita akan membahasnya lebih lanjut nanti.

Mari kita tinjau secara singkat mekanisme jaringan saraf. Proses berikut dilakukan di dalam sebuah node jaringan saraf:

1. Jumlah berbobot dari sinyal masukan dihitung.

$$
v = w_1 x_1 + w_2 x_2 + w_3 x_3 + b
$$

$$
= wx + b
$$

2. Keluaran dari fungsi aktivasi terhadap jumlah berbobot diteruskan ke luar.

$$
y = \varphi(v) = \varphi(wx + b)
$$

## **Lapisan Jaringan Saraf** 

Sebagaimana otak merupakan jaringan raksasa dari neuron, jaringan saraf (neural network) adalah jaringan dari node. Berbagai jenis jaringan saraf dapat dibuat tergantung bagaimana node-node tersebut terhubung. Salah satu jenis jaringan saraf yang paling umum digunakan menerapkan struktur berlapis dari node seperti ditunjukkan pada Gambar 2-3.

![Gambar 2-3.png](<attachment:Gambar 2-3.png>)

**Gambar 2-3.** Struktur berlapis dari node

Kelompok node berbentuk kotak pada Gambar 2-3 disebut input layer. Node pada input layer hanya berfungsi sebagai jalur yang menyalurkan sinyal masukan ke node berikutnya. Oleh karena itu, mereka tidak menghitung jumlah berbobot (weighted sum) maupun fungsi aktivasi. Inilah alasan mengapa mereka digambarkan dengan kotak dan dibedakan dari node lain yang berbentuk lingkaran. Sebaliknya, kelompok node di paling kanan disebut output layer. Keluaran dari node-node ini menjadi hasil akhir dari jaringan saraf. Lapisan-lapisan di antara input layer dan output layer disebut hidden layers. Mereka diberi nama demikian karena tidak dapat diakses dari luar jaringan saraf.

Jaringan saraf telah berkembang dari arsitektur sederhana menuju struktur yang lebih kompleks. Awalnya, pelopor jaringan saraf hanya membuat arsitektur dengan input layer dan output layer, yang disebut jaringan saraf lapisan tunggal (single-layer neural networks). Ketika hidden layer ditambahkan ke jaringan saraf lapisan tunggal, hal itu menghasilkan jaringan saraf multi-lapisan (multi-layer neural network). Dengan demikian, jaringan saraf multi-lapisan terdiri atas input layer, satu atau lebih hidden layer, dan output layer.

Jaringan saraf dengan satu hidden layer disebut shallow neural network atau jaringan saraf dangkal. Sedangkan jaringan saraf multi-lapisan dengan dua atau lebih hidden layer disebut deep neural network. Sebagian besar jaringan saraf modern yang digunakan dalam aplikasi praktis merupakan deep neural networks.

Jaringan saraf berawal dari single-layer neural network kemudian berkembang menjadi shallow neural network, dan akhirnya menuju deep neural network.

Jaringan saraf dalam telah mulai mendapat sorotan serius sejak pertengahan tahun 2000-an, setelah dua dekade berlalu sejak pengembangan shallow neural network. Untuk waktu yang lama, istilah jaringan saraf multi-lapisan hanya merujuk pada jaringan dengan satu hidden layer. ## **Pembelajaran *Supervised* Jaringan Saraf**

  Bagian ini memperkenalkan konsep dan proses pembelajaran *supervised* dari jaringan saraf.  Hal ini dibahas dalam bagian "Jenis Pembelajaran Mesin" pada Bab 1. Dari banyak metode pelatihan, buku ini hanya membahas pembelajaran *supervised*. Oleh karena itu, hanya pembelajaran *supervised* yang dibahas untuk jaringan saraf. Secara umum, pembelajaran *supervised* jaringan saraf dilakukan dalam langkah-langkah berikut:

  1. Inisialisasi bobot dengan nilai yang memadai.

  2. Ambil "*input*" dari data pelatihan, yang diformat sebagai *{input, output yang benar}*, dan masukkan ke dalam jaringan saraf.  Dapatkan keluaran dari jaringan saraf dan hitung kesalahan dari keluaran yang benar.

  3. Sesuaikan bobot untuk mengurangi kesalahan.

  4. Ulangi Langkah 2-3 untuk semua data pelatihan.

  Langkah-langkah ini pada dasarnya identik dengan proses pembelajaran *supervised* pada bagian "Jenis-jenis Pembelajaran Mesin".  Ini masuk akal karena pelatihan pembelajaran *supervised* adalah proses yang memodifikasi model untuk mengurangi perbedaan antara keluaran yang benar dan keluaran model.  Satu-satunya perbedaan adalah bahwa modifikasi model menjadi perubahan bobot untuk jaringan saraf.  Gambar 2-10 mengilustrasikan konsep pembelajaran *supervised* yang telah dijelaskan sejauh ini. ## **Pelatihan Jaringan Saraf Satu Lapis: Aturan Delta**

Seperti yang telah dibahas sebelumnya, jaringan saraf menyimpan informasi dalam bentuk bobot. Oleh karena itu, untuk melatih jaringan saraf dengan informasi baru, bobot harus diubah sesuai dengan itu.  Pendekatan sistematis untuk memodifikasi bobot sesuai dengan informasi yang diberikan disebut aturan pembelajaran.  Karena pelatihan adalah satu-satunya cara bagi jaringan saraf untuk menyimpan informasi secara sistematis, aturan pembelajaran adalah komponen penting dalam penelitian jaringan saraf.
  
Pada bagian ini, kita membahas aturan delta, aturan pembelajaran representatif dari jaringan saraf lapisan tunggal. Meskipun tidak mampu untuk pelatihan jaringan saraf multi-lapisan, aturan ini sangat berguna untuk mempelajari konsep penting dari aturan pembelajaran jaringan saraf.

Pertimbangkan sebuah jaringan saraf lapisan tunggal, seperti yang ditunjukkan pada Gambar 2-11. Pada gambar tersebut, *$d_i$* adalah keluaran yang benar dari node keluaran *i*.

![Gambar 2-11.png](<attachment:Gambar 2-11.png>)

**Gambar 2-11.** Sebuah jaringan saraf lapisan tunggal

Singkatnya, aturan delta menyesuaikan bobot dengan algoritme berikut:

â€œJika sebuah node input berkontribusi terhadap kesalahan pada node keluaran, maka bobot antara kedua node tersebut disesuaikan sebanding dengan nilai input *$x_j$* dan kesalahan keluaran *$e_i$*.â€

</div>

<div align="justify">

Aturan ini dapat diekspresikan dalam persamaan sebagai:

$$
w_{ij} \leftarrow w_{ij} + \alpha e_i x_j \quad \tag{Persamaan 2.2}
$$

**Dimana:**

$x_j$ = Output dari node input $j, (j=1, 2, 3)$

$e_i$ = Kesalahan pada node output $i$

$w_ij$ = Berat antara node output i dan node input $j$

$a$ =  Laju pembelajaran $(0 < a \leq 1)$

Laju proses, Î±, menentukan seberapa besar perubahan bobot per waktu. Jika nilai ini terlalu tinggi, output akan berfluktuasi di sekitar solusi dan gagal konvergen. Sebaliknya, jika terlalu rendah, perhitungan mencapai solusi terlalu lambat. 

Sebagai contoh konkret, pertimbangkan jaringan saraf tunggal lapisan, yang terdiri dari tiga node input dan satu node output, seperti yang ditunjukkan pada Gambar 2-12. Untuk kemudahan, kita mengasumsikan tidak ada bias untuk node output pada saat ini. Kita menggunakan fungsi aktivasi linier; yaitu, jumlah tertimbang langsung ditransfer ke output.

![Gambar 2-12.png](<attachment:Gambar 2-12.png>)

**Gambar 2-12.** Jaringan saraf tiruan berlapis tunggal dengan tiga node input dan satu node output

Perhatikan bahwa angka pertama pada subscript (1) menunjukkan nomor node tempat input masuk. Misalnya, bobot antara node input 2 dan node output 1 dilambangkan sebagai w12. Notasi ini memudahkan operasi matriks; bobot yang terkait dengan node i dialokasikan pada baris ke-i darimatriks bobot. Menerapkan aturan delta Persamaan 2.2 pada jaringan saraf tiruan contoh ini menghasilkan pembaruan bobot sebagai:

$$
w_{11} \leftarrow w_{11} + \alpha e_1 x_1
$$

$$
w_{12} \leftarrow w_{12} + \alpha e_1 x_2
$$

$$
w_{13} \leftarrow w_{13} + \alpha e_1 x_3
$$

Mari kita ringkas proses pelatihan menggunakan aturan delta untuk jaringan neural tunggal.
1. Inisialisasi bobot pada nilai yang memadai.
2. Ambil â€œinputâ€ dari data pelatihan { input, output benar } dan masukkan ke dalam jaringan neural. Hitung kesalahan output, yi, dari output benar,di, terhadap input.
$$
e_i = d_i - y_i
$$

3. Hitung pembaruan berat sesuai dengan aturan delta berikut:
$$
\Delta w_{ij} = \alpha e_i x_j
$$

4. Sesuaikan bobot sebagai berikut:
$$
w_{ij} \leftarrow w_{ij} + \Delta w_{ij}
$$
5. Lakukan Langkah 2-4 untuk semua data pelatihan.
6. Ulangi Langkah 2-5 hingga kesalahan mencapai tingkat toleransi yang dapat diterima.

Langkah-langkah ini hampir identik dengan proses pembelajaran terawasi dalam bagian â€œPembelajaran Terawasi Jaringan Saraf Tiruanâ€. Perbedaan satu-satunya adalah penambahan Langkah 6. Langkah 6 hanya menyatakan bahwa seluruh proses pelatihan diulang. Setelah Langkah 5 selesai, model telah dilatih dengan setiap titik data. Lalu, mengapa kita melatihnya menggunakan
semua data pelatihan yang sama? Ini karena aturan delta mencari solusi saat mengulang proses, bukan menyelesaikannya sekaligus. Seluruh proses diulang, karena
melatih ulang model dengan data yang sama dapat meningkatkan model. Sebagai referensi, jumlah iterasi pelatihan, di mana setiap data pelatihan
melalui Langkah 2-5 sekali, disebut epoch. Misalnya, epoch = 10 berarti jaringan saraf melalu 10 proses pelatihan berulang dengan dataset yang sama.

Aturan delta adalah jenis metode numerik yang disebut gradien turun. Gradien
turun dimulai dari nilai awal dan berlanjut menuju solusi. Namanya berasal dari
perilakunya yang mencari solusi seolah-olah bola menggelinding turun bukit sepanjang
jalur tercuram. Dalam analogi ini, posisi bola adalah output acak dari
model, dan dasar bukit adalah solusi. Perlu dicatat bahwa metode gradien turun

## Generalized Delta Rule
### (Aturan Delta yang Digeneralisasi)

Aturan delta yang dijelaskan sebelumnya kini dianggap kurang relevan (usang) karena hanya berlaku untuk fungsi aktivasi linear. Penelitian terbaru menunjukkan bahwa ada bentuk yang lebih umum (generalisasi) dari aturan delta yang dapat digunakan untuk berbagai jenis fungsi aktivasi. Delta Rule adalah salah satu algoritma pembelajaran penting pada jaringan saraf tiruan (Neural Network) untuk memperbarui bobot (weights). Versi generalized dari aturan ini digunakan agar bisa diterapkan pada berbagai fungsi aktivasi, tidak hanya fungsi linear.

**Generalized Delta Rule** adalah aturan pembaruan bobot (weight update rule) yang digunakan dalam **pembelajaran jaringan saraf tiruan (Artificial Neural Networks)**.  
Versi ini lebih umum dibandingkan *delta rule klasik*, karena bisa diterapkan pada berbagai **fungsi aktivasi** (activation function), bukan hanya fungsi linear.

### Rumus Umum Generalized Delta Rule
Aturan pembaruan bobot ditulis sebagai:

$$
w_{ij} \leftarrow w_{ij} + \alpha \, \delta_i \, x_j
$$

**Keterangan:**
- $w_{ij}$ : bobot dari neuron input ke-$j$ menuju neuron output ke-$i$  
- $\alpha$ : learning rate (kecepatan belajar)  
- $\delta_i$ : error term (kesalahan pada neuron ke-$i$)  
- $x_j$ : nilai input dari neuron ke-$j$  

### Definisi $\delta_i$ (Error Term)
Error term didefinisikan sebagai:

$$
\delta_i = e_i \cdot \phi'(v_i)
$$

**Keterangan:**
- $e_i$ : error dari output node $i$  
- $v_i$ : jumlah tertimbang (weighted sum) dari input pada node $i$  
- $\phi'(v_i)$ : turunan dari fungsi aktivasi $\phi(v)$ pada node $i$  

### Kasus Fungsi Aktivasi Linear
Jika fungsi aktivasi adalah fungsi linear:

$$
\phi(x) = x
$$
maka turunannya adalah:
$$
\phi'(x) = 1
$$
Sehingga:
$$
\delta_i = e_i \cdot 1 = e_i
$$
Maka persamaan bobot kembali menjadi bentuk **Delta Rule klasik**:
$$
w_{ij} \leftarrow w_{ij} + \alpha \, e_i \, x_j
$$
 Artinya, *delta rule klasik hanya berlaku untuk fungsi aktivasi linear*.

### Kasus Fungsi Aktivasi Sigmoid

Fungsi aktivasi sigmoid didefinisikan sebagai:

$$
\phi(x) = \frac{1}{1 + e^{-x}}
$$

Turunan fungsi sigmoid adalah:

$$
\phi'(x) = \phi(x) \cdot \big(1 - \phi(x)\big)
$$

Dengan demikian, error term untuk neuron $i$ menjadi:

$$
\delta_i = e_i \cdot \phi(v_i) \cdot \big(1 - \phi(v_i)\big)
$$
untuk ilustrasinya dapat dilihat pada gambar bberikut:

![Gambar 2-14.png](<attachment:Gambar 2-14.png>)

**Gambar 2-14.** fungsi sigmoid yang didefinisikan

### Generalized Delta Rule dengan Fungsi Sigmoid

Jika kita substitusikan $\delta_i$ pada rumus update bobot, maka diperoleh:

$$
w_{ij} \leftarrow w_{ij} + \alpha \, e_i \, \phi(v_i) \cdot \big(1 - \phi(v_i)\big) \, x_j
$$

**Interpretasi:**
- Bobot diperbarui **sebanding dengan error output** ($e_i$),  
- Dikalikan dengan **sensitivitas fungsi aktivasi** ($\phi(v_i)(1-\phi(v_i))$),  
- Dan juga dengan **nilai input** ($x_j$).

### Inti Konsep

Walaupun rumus untuk fungsi sigmoid terlihat lebih rumit dibandingkan dengan linear, **konsep dasarnya tetap sama**:
ialah Bobot ($w_{ij}$) diperbarui berdasarkan **besar error output** dan **kontribusi input**.  

### Ringkasan Persamaan

| Fungsi Aktivasi | Turunan $\phi'(x)$ | $\delta_i$ (Error Term) | Update Bobot |
|------------------|-------------------|-------------------------|---------------|
| **Linear** $\phi(x)=x$ | $1$ | $\delta_i = e_i$ | $w_{ij} \leftarrow w_{ij} + \alpha e_i x_j$ |
| **Sigmoid** $\phi(x)=\frac{1}{1+e^{-x}}$ | $\phi(x)(1-\phi(x))$ | $\delta_i = e_i \cdot \phi(v_i)(1-\phi(v_i))$ | $w_{ij} \leftarrow w_{ij} + \alpha e_i \phi(v_i)(1-\phi(v_i)) x_j$ |

### Kesimpulan

1. **Delta rule klasik** hanya berlaku untuk fungsi linear.  
2. **Generalized delta rule** memungkinkan penggunaan fungsi aktivasi non-linear seperti sigmoid.  
3. Perubahan bobot ($w_{ij}$) selalu dipengaruhi oleh:  
   - besar error output ($e_i$),  
   - turunan fungsi aktivasi pada node ($\phi'(v_i)$),  
   - nilai input ($x_j$).

````python
import numpy as np

# Sigmoid function
def sigmoid(v):
    return 1 / (1 + np.exp(-v))

# Delta rule for SGD
def DeltaSGD(W, X, D, alpha=0.1, epochs=1000):
    for epoch in range(epochs):
        for x, d in zip(X, D):
            # Forward pass
            v = np.dot(W, x)
            y = sigmoid(v)
            
            # Error
            e = d - y
            
            # Delta (Eq. 2.7)
            delta = y * (1 - y) * e
            
            # Weight update
            dW = alpha * delta * x
            W = W + dW
    return W

# Training data (dengan bias term = 1 di paling depan)
X = np.array([
    [1, 0, 0, 1],  # bias + inputs
    [1, 0, 1, 1],
    [1, 1, 0, 1],
    [1, 1, 1, 1]
])
D = np.array([0, 0, 1, 1])

# Inisialisasi bobot random
np.random.seed(42)  # biar hasil sama setiap run
W = np.random.randn(4)

# Latih dengan SGD
W_trained = DeltaSGD(W, X, D, alpha=0.1, epochs=1000)
print("Trained Weights:", W_trained)

# Uji hasil training
print("\nTesting after training:")
for x, d in zip(X, D):
    y = sigmoid(np.dot(W_trained, x))
    print(f"Input: {x[1:]} Target: {d} Output: {y:.4f}")
````

    alpha = 0.9;

```matlab
function W = DeltaBatch(W, X, D)
    alpha = 0.9;

    dwsum = zeros(3, 1);

    N = 4;
    for k = 1:N
        x = X(k, :)';
        d = D(k);

        v = W*x;
        y = Sigmoid(v);

        e = d - y;
        delta = y*(1-y)*e;

        dw = alpha*delta*x;

        dwsum = dwsum + dw;
    end
    dwavg = dwsum / N;
```

```
        y1 = Sigmoid(v1);
        es1 = es1 + (d - y1)^2;
        v2 = W2*x;
        y2 = Sigmoid(v2);
        es2 = es2 + (d - y2)^2;
    end
    E1(epoch) = es1 / N;
    E2(epoch) = es2 / N;
end
plot(E1, 'r')
hold on
plot(E2, 'b:')
xlabel('Epoch')
ylabel('Average of Training error')
legend('SGD', 'Batch')
```

## **Kesimpulan**

Bab ini membahas konsep-konsep berikut:

â€¢ Jaringan saraf (neural network) adalah jaringan node yang meniru neuron otak. Node menghitung jumlah berbobot dari sinyal masukan dan menghasilkan keluaran melalui fungsi aktivasi.

â€¢ Sebagian besar jaringan saraf dibangun dengan node berlapis. Pada jaringan saraf berlapis, sinyal masuk melalui lapisan input, diteruskan ke lapisan tersembunyi (hidden layer), lalu keluar melalui lapisan output.

â€¢ Dalam praktiknya, fungsi linear tidak dapat digunakan sebagai fungsi aktivasi di lapisan tersembunyi, karena fungsi linear menghilangkan efek lapisan tersembunyi. Namun, pada beberapa masalah seperti regresi, node di lapisan output dapat menggunakan fungsi linear.

â€¢ Pada jaringan saraf, supervised learning menerapkan proses untuk menyesuaikan bobot dan mengurangi perbedaan antara keluaran yang benar dengan keluaran dari jaringan saraf (Gambar 2-26).

![Gambar 2-26.png](<attachment:Gambar 2-26.png>)

**Gambar 2-26** Tinjauan tentang supervised learning

â€¢ Metode yang digunakan untuk menyesuaikan bobot sesuai dengan data pelatihan disebut learning rule.

â€¢ Ada tiga jenis utama perhitungan error: stochastic gradient descent (SGD), batch, dan mini-batch.

â€¢ Delta rule adalah learning rule yang paling mewakili dalam jaringan saraf. Rumusnya dapat berbeda-beda tergantung pada fungsi aktivasi yang digunakan.

$$ \delta_i = \varphi'(v_i) e_i $$

$$
w_{ij} \leftarrow w_{ij} + \alpha \delta_i x_j
$$

â€¢ Delta rule adalah metode iteratif yang secara bertahap mencapai solusi. Karena itu, jaringan harus dilatih berulang kali dengan data pelatihan hingga error berkurang ke tingkat yang memuaskan.

â€¢ Single-layer neural network hanya dapat diterapkan pada jenis masalah tertentu, sehingga penggunaannya sangat terbatas. Untuk mengatasi keterbatasan mendasar ini, dikembangkan multi-layer neural network.
